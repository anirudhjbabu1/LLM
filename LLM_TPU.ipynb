```python
import os
import sys
import jax
import nest_asyncio
from absl import logging as absl_logging

absl_logging.set_verbosity(absl_logging.INFO)
nest_asyncio.apply()

# 1. Block MaxText from searching for the broken Transformer Engine
os.environ["ENABLE_TE_JAX"] = "0"
os.environ["MAXTEXT_DISABLE_TE"] = "1"

# 2. Enforce repository directory tracking context
if os.path.exists("/content/maxtext"):
    os.chdir("/content/maxtext")

# 3. Hard-purge references from background memory
for mod in list(sys.modules.keys()):
    if any(pkg in mod for pkg in ["maxtext", "transformers", "drjax"]):
        del sys.modules[mod]

# 4. Force-insert absolute local source path
LOCAL_SRC = "/content/maxtext/src"
if LOCAL_SRC in sys.path:
    sys.path.remove(LOCAL_SRC)
sys.path.insert(0, LOCAL_SRC)

# Re-define the missing variable!
MAXTEXT_CONFIGS_DIR = "/content/maxtext/src/maxtext/configs"

# Hardware Auto-Detection
ACTIVE_BACKEND = jax.default_backend()
print(f"✅ Hardware Accelerator Detected: {ACTIVE_BACKEND.upper()}")
print(f"✅ Available Devices: {jax.devices()}")
```

```python
from maxtext.trainers.pre_train import train

argv_base = [
    "",
    f"{MAXTEXT_CONFIGS_DIR}/base.yml",
    "model_name=qwen3-0.6b",
    "steps=50",
    "dataset_type=synthetic",
    "enable_checkpointing=false",
    "base_output_directory=/tmp/maxtext_qwen06",
    "run_name=qwen_06b_base_profile",
    "per_device_batch_size=1",
    f"hardware={ACTIVE_BACKEND}",
    "attention=dot_product",
    "max_target_length=512"  # 👈 Keeps us under the 16GB VRAM limit
]

print(f"🚀 Executing Run A (0.6B Base) on {ACTIVE_BACKEND.upper()} (VRAM Optimized)...")
train.main(argv_base)
```
